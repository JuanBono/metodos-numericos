# Interpolación

## Polinomio Interpolador de Lagrange

### Descripción:
Dado un conjunto de k + 1 puntos

$$ (x_0, y_0),\ldots,(x_k, y_k) $$ 

donde todos los xj se asumen distintos, el polinomio interpolador en la forma de Lagrange es la combinación lineal

$$ L(x) = \sum_{j=0}^{k} y_j \ell_j(x) $$

de bases polinómicas de Lagrange

$$ \ell_j(x) = \prod_{i=0,\, i\neq j}^{k} \frac{x-x_i}{x_j-x_i} = \frac{x-x_0}{x_j-x_0}\cdots \frac{x-x_{j-1}}{x_j-x_{j-1}}\frac{x-x_{j+1}}{x_j-x_{j+1}}\cdots \frac{x-x_{k}}{x_j-x_{k}} $$


### Ventajas:
Su implementacion es muy sencilla, convirtiendolo en un método muy conveniente si se dispone de un set reducido de datos.

### Desventajas:
Es efectivo para la mayoria de las funciones hasta polinomios de grado 6. Si se dispone de mas de 6 puntos, es mejor usar otro metodo como Spline Cúbico. 
Ademas es muy costoso computacionalmente y a mayor cantidad de datos, mayores son las oscilaciones entre puntos consecutivos
.
### Implementación



```julia
function polinomio_de_lagrange(px, puntos)
  n = length(puntos[1]) 
  L = Real[]
  for j = 1:n # Armo lista con los valores de los polinomios fundamentales de lagrange valuados en px
    Lj = 1
    for k = 1:n
        if (j != k) 
          Lj *= (px - puntos[1][k]) / (puntos[1][j] - puntos[1][k])
        end
    end
    push!(L, Lj)
  end

  py = 0 # polinomio de lagrange valuado en px
  for i = 1:n
    py += puntos[2][i] * L[i]
  end

  return py
    
end
```

### Caso de Prueba
Se desea calcular el valor, segun el polinomio interpolador de Lagrange, de $ f(x) = tan(x) $, para ello se cuenta con una serie de puntos resultado de su evaluacion en distintos valores y un punto especifico $ p_{x} $
$$ \begin{align*} 
 x_{0}&=-1.5 & f(x_{0})&=-14.1014 \\
 x_{1}&=-0.75 & f(x_{1})&=-0.931596 \\
 x_{2}&=0 & f(x_{2})&= 0 \\
 x_{3}&=0.75, & f(x_{3})&= 0.931596 \\
 x_{4}&=1.5, & f(x_{4})&= 14.1014 \\
 \end{align*} $$


```julia
x0 = -1.5
x1 = -0.75
x2 = 0
x3 = 0.75
x4 = 1.5

xs = [x0, x1, x2, x3, x4]
ys = map(tan, xs)

"Valores Inicializados"
```

```julia
polinomio_de_lagrange(0.5, (xs, ys))
```

```julia
using Gadfly
set_default_plot_size(20cm, 15cm)
lagrangeParaGraficar(x) = polinomio_de_lagrange(x, (xs, ys))
plot([tan,lagrangeParaGraficar],-1.5,1.5)
```
